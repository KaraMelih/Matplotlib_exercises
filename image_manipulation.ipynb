## Using numpy to create 2-D objects

```python
x = np.linspace(1,5, 100)
y = np.linspace(1,5, 50)
# generate 2 2d grids for the x & y bounds
yy, xx = np.meshgrid(x,y)
z = np.sin(xx)**10 + np.cos(10 + yy*xx) * np.cos(xx)
plt.imshow(z);
```

Let's take a look at an example image

```python
ima = plt.imread('kodluyoruz.jpg')
# show metadata of ima
print(type(ima), ima.dtype, ima.ndim, ima.shape)
```

```python
plt.imshow(ima)
# plt.axis('off')
rocket = ima[55:105,160:190]
print(rocket.shape)
```

```python
plt.imshow(rocket)
```

```python
# Lets flip this
rocket_flipped = rocket[::-1,:,:]  # revert x-axis
plt.imshow(rocket_flipped)
```

```python
# swap axis
rocket_swap = np.swapaxes(rocket, 1, 0)
plt.imshow(rocket_swap)
```

```python
# rockets sizes 
dx, dy, _ = rocket.shape

# make a copy and modify the original image
ima_new = ima.copy()
# called fancy-indexing
ima_new[20:20+dx,10:10+dy,:] = rocket
```

```python
plt.imshow(ima_new)
```

```python
ima_new[60:60+dy,50:50+dx,:] = rocket_swap[:,::-1,:]
plt.imshow(ima_new)
```
